```
# Jawaban Latihan dari PDF Desain Analisis Algoritma
# Semua latihan digabungkan dalam satu file.

# Import library yang dibutuhkan untuk latihan DataFrame dan Matrix
import pandas as pd
import numpy as np
from datetime import date

print("--- Pengerjaan Latihan Struktur Data ---")

# ====================================================================
print("\n### Latihan Halaman 6: Membuat List Mahasiswa ###")
# Menggunakan tanggal hari ini secara otomatis
tanggal_hari_ini = date.today().strftime("%d %B %Y")
mahasiswa_list = [
    "Lathifah Alfat",
    "2023071001",
    "Informatika",
    "Desain Analisis Algoritma",
    tanggal_hari_ini,
    "Universitas Pembangunan Jaya"
]
print("Isi list 'mahasiswa_list':")
print(mahasiswa_list)

# ====================================================================
print("\n### Latihan Halaman 8: Mencetak NIM dengan Indeks ###")
# Indeks NIM adalah 1 (elemen kedua)
nim = mahasiswa_list[1]
print(f"NIM dari list: {nim}")

# ====================================================================
print("\n### Latihan Halaman 9: Mencetak Nama Universitas dengan Indeks ###")
# Indeks Universitas adalah 5 (elemen keenam)
universitas = mahasiswa_list[5]
print(f"Universitas dari list: {universitas}")

# ====================================================================
print("\n### Latihan Halaman 11: Mencetak NIM dan Prodi dengan Slicing ###")
# Mengambil elemen dari indeks 1 sampai sebelum indeks 3
nim_dan_prodi = mahasiswa_list[1:3]
print("NIM dan Prodi (slicing):")
print(nim_dan_prodi)

# ====================================================================
print("\n### Latihan Halaman 13: Iterasi List dan Menambahkan Kata ###")
print("Isi list dengan tambahan 'UPJ':")
for item in mahasiswa_list:
    print(f"UPJ {item}")

# ====================================================================
print("\n### Latihan Halaman 15: Membuat Tuple ###")
UPJ_tuple = ("Universitas", "Pembangunan", "Jaya")
print("Isi tuple 'UPJ_tuple':")
print(UPJ_tuple)

# ====================================================================
print("\n### Latihan Halaman 16: Membuat Tuple Bersarang ###")
nested_tuple = (
    100,
    (200, 400, 600),
    300,
    (400, 800)
)
print("Isi 'nested_tuple':")
print(nested_tuple)

# ====================================================================
print("\n### Latihan Halaman 20: Membuat Dictionary ###")
mahasiswa_dict = {
    "Nama": "Lathifah Alfat",
    "NIM": "2023071001",
    "Prodi": "Informatika",
    "Universitas": "UPJ"
}
print("Isi dictionary 'mahasiswa_dict':")
print(mahasiswa_dict)

# ====================================================================
print("\n### Latihan Halaman 23: Operasi pada Sets ###")
# Berdasarkan diagram Venn pada slide
yellow = {'dandelions', 'leaves', 'fire hydrant'}
red = {'rose', 'blood', 'leaves', 'fire hydrant'}

# Gabungan (Union)
gabungan = yellow.union(red)
print(f"Hasil gabungan (yellow | red): {gabungan}")

# Irisan (Intersection)
irisan = yellow.intersection(red)
print(f"Hasil irisan (yellow & red): {irisan}")

# ====================================================================
print("\n### Latihan Halaman 28: Membuat DataFrame ###")
data = {
    'No': [1, 2, 3],
    'Prodi': ['Informatika', 'Sistem Informasi', 'Teknik Sipil'],
    'Mahasiswa': [50, 55, 40],
    'Laki-laki': [30, 30, 30],
    'Perempuan': [20, 25, 10]
}
mhs_df = pd.DataFrame(data)
print("Isi DataFrame 'mhs_df':")
print(mhs_df)

# ====================================================================
print("\n### Latihan Halaman 30: Membuat dan Transpose Matrix ###")
# Membuat matrix dengan NumPy
myMatrix = np.array([
    [100, 200, 300],
    [700, 600, 500],
    [900, 1000, 800]
])
print("\nMatrix Awal:")
print(myMatrix)

# Melakukan transpose pada matrix
transposed_matrix = myMatrix.T
print("\nMatrix Setelah di-Transpose:")
print(transposed_matrix)

print("\n--- Selesai ---")
```
